## Q1. 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
no_10 = df[df['smoker'] == 'no']['charges'].quantile(0.9)
```

```python
no_mean = df[(df['smoker'] == 'no') & (df['charges'] >= no_10)]['charges'].mean()
```

```python
yes_10 = df[df['smoker'] == 'yes']['charges'].quantile(0.9)
```

```python
yes_mean = df[(df['smoker'] == 'yes') & (df['charges'] >= yes_10)]['charges'].mean()
```

```python
yes_mean - no_mean
```

## Q1. bedrooms 의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10%값의 차이를 구하여라

```python
top = df['bedrooms'].value_counts().reset_index()['index'][0]
```

```python
df[df['bedrooms'] == top]['price'].quantile(0.9) - df[df['bedrooms'] == top]['price'].quantile(0.1)
```
