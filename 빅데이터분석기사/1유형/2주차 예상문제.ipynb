## Q2. 40대(연령대코드 40,45) 여성 중 '총콜레스테롤'값이 40대 여성의 '총콜레스테롤' 중간값 이상을  가지는 그룹과 50대(연령대코드 50,55) 여성 중 '총콜레스테롤'값이 50대 여성의 '총콜레스테롤' 중간값 이상을 가지는 두 그룹의 '수축기혈압'이 독립성,정규성,등분산성이 만족하는것을 확인했다. 두 그룹의 '수축기혈압'의 독립표본 t 검증 결과를 통계값, p-value 구분지어 구하여라.

```python
f1 = df[(df['성별코드'] == 'F') & (df['연령대코드(5세단위)'].isin([40,45]))]
f2 = df[(df['성별코드'] == 'F') & (df['연령대코드(5세단위)'].isin([50,55]))]
```

```python
p1 = f1[f1['총콜레스테롤'] >= f1['총콜레스테롤'].median()]['수축기혈압']
p2 = f2[f2['총콜레스테롤'] >= f2['총콜레스테롤'].median()]['수축기혈압']
```

```python
import scipy.stats
scipy.stats.ttest_ind(p1, p2, equal_var = True)
```

## Q1. 수축기혈압과 이완기 혈압기 수치의 차이를 새로운 컬럼('혈압차') 으로 생성하고, 연령대 코드별 각 그룹 중 '혈압차' 의 분산이 5번째로 큰 연령대 코드를 구하여라

```python
df['혈압차'] = df['수축기혈압'] - df['이완기혈압']
```

```python
df.groupby(['연령대코드(5세단위)'])['혈압차'].var().sort_values()
```

## Q2. 비만도를 나타내는 지표인 WHtR는 허리둘레 / 키로 표현한다. 일반적으로 0.58이상이면 비만으로 분류한다. 데이터중 WHtR 지표상 비만인 인원의 남/여 비율을 구하여라

```python
df['WHtR'] = df['허리둘레'] / df['신장(5Cm단위)']
len(df[(df['성별코드'] == 'M') & (df['WHtR'] >= 0.58)]) / len(df[(df['성별코드'] == 'F') & (df['WHtR'] >= 0.58)])
```
