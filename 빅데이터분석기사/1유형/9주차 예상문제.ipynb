```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")
```

```python
df
```

## Q1. 매년 5월달의 open 가격의 평균값을 데이터 프레임으로 표현하라

```python
df['Date'] = pd.to_datetime(df['Date'])
df[df['Date'].dt.month == 5].groupby(df['Date'].dt.year)['Open'].mean().to_frame()
```
