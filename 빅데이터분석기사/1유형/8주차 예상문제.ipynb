## Q1. pose값에 따른 각 motion컬럼의 중간값의 가장 큰 차이를 보이는 motion컬럼은 어디이며 그값은?

```python
df
```

## 정보(row수)가 가장 많은 상위 3차종의 price값의 각 평균값은?

```python
model_list = df.groupby('model').size().sort_values().index[-3:].to_list()
```

```python
model_list
```

```python
df[df['model'].isin(model_list)].groupby('model').mean()['price']
```
