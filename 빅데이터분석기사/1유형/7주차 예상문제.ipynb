## Q1. 남성들의 연령대별 (10살씩 구분 0-9세 10-19세 ...) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

```python
men_df = df[df['Sex'] == 'M']
```

```python
men_df['Age2'] = men_df['Age'] // 10 * 10
```

```python
men_df
```

```python
men_df.groupby('Age2').mean()['Na_to_K'].to_frame()
```
