## Q3. Channel 데이터중 2021-10-03일 이후의 각 채널의 처음 구독자 수(subcnt)를 출력하라

```python
channel[channel['ct'] >= '2021-10-03'].sort_values(['ct', 'channelid']).drop_duplicates('channelname', keep='first')[['channelname', 'subcnt']]
```

## Q4. 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
a = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-10-03 03'][['channelname','subcnt']]
b = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-11-01 15'][['channelname','subcnt']]
a.sort_values(by = 'channelname', inplace = True)
b.sort_values(by = 'channelname', inplace = True)
a.reset_index(inplace = True)
b.reset_index(inplace = True)
```

```python
a
```

```python
b
```

```python
inc = b.subcnt - a.subcnt
a['inc'] = inc
a[['channelname','inc']]
```

## Q5. 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다. 공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라

```python
import datetime

ep_one = video.loc[video.videoname.str.contains('1')].sort_values('ct').reset_index(drop=True)

ep_one[
        (ep_one.ct.diff(1) >= datetime.timedelta(minutes=20)) | \
        (ep_one.ct.diff(1) <= datetime.timedelta(minutes=5))
      ]
```

## Q7. "공범" 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.  공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

```python
video['time'] = video['ct'].dt.hour
video[video['time'] == 21].sort_values(['videoname', 'ct']).\
drop_duplicates('videoname').sort_values('viewcnt', ascending=False)[['videoname', 'viewcnt', 'ct']].reset_index(drop=True)
```

## Q8. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

```python
Q8_df = video.sort_values('ct').drop_duplicates('videoname', keep='last')
Q8_df['ratio'] = Q8_df['dislikecnt'] / Q8_df['likecnt']

Q8_df.sort_values('ratio')[['videoname', 'ratio']].reset_index(drop=True)
```
