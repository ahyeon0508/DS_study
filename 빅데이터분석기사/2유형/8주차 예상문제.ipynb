# 8주차 예상문제 1

데이터 출처 :https://www.kaggle.com/kyr7plus/emg-4(참고, 데이터 수정)

데이터 설명 : 센서데이터로 동작 유형 분류

평가지표 : f1_scoe

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/test.csv

출처 : 8주차 예상문제 1 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

## 데이터 분리

```python
from sklearn.model_selection import train_test_split

X = train.drop(['pose'], axis=1)
y = train['pose']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

# 모델 평가

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```

## 테스트

```python
submission = pd.DataFrame(model.predict(test))
submission.to_csv("submission_8주차_1.csv", index=False, header=False)
```

# 8주차 예상문제 2

데이터 출처 :https://www.kaggle.com/mysarahmadbhat/hyundai-used-car-listing(참고, 데이터 수정)

데이터 설명 : 현대차 스펙에 따른 가격 예측문제

평가지표 : r2_score

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/test.csv

출처 : 8주차 예상문제 2 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

## 데이터 분리

```python
from sklearn.model_selection import train_test_split

X = train.drop(['price'], axis=1)
y = train['price']

X = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

# 모델 평가

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```

## 테스트

```python
test = pd.get_dummies(test)
submission = pd.DataFrame(model.predict(test))
submission.to_csv("submission_8주차_2.csv", index=False, header=False)
```
