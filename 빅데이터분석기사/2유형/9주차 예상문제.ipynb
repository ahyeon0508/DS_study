# 9주차 예상문제 1

데이터 출처 :https://www.kaggle.com/pritsheta/diabetes-dataset(참고, 데이터 수정)

데이터 설명 : 당뇨여부 판단하기

평가지표 : f1_score

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/test.csv

출처 : 9주차 예상문제 1 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

## 데이터 분리

```python
from sklearn.model_selection import train_test_split

X = train.drop(['Outcome'], axis=1)
y = train['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

# 모델 평가

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```

## 테스트

```python
submission = pd.DataFrame(model.predict(test))
submission.to_csv("submission_9주차_1.csv", index=False, header=False)
```

# 9주차 예상문제 2

데이터 출처 :https://www.kaggle.com/sansuthi/dry-bean-dataset

(참고, 데이터 수정)

데이터 설명 : 콩 종류 분류

평가지표 : accuracy

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/bean/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/bean/test.csv

출처 : 9주차 예상문제 2 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/bean/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/bean/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

## 데이터 분리

```python
from sklearn.model_selection import train_test_split

X = train.drop(['Class'], axis=1)
y = train['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

# 모델 평가

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```

## 테스트

```python
submission = pd.DataFrame(model.predict(test))
submission.to_csv("submission_9주차_2.csv", index=False, header=False)
```
