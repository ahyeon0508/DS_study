# 5주차 예상문제 1

데이터 출처 :https://www.kaggle.com/mirichoi0218/insurance/code(참고, 데이터 수정)

데이터 설명 : 의료비용 예측문제

평가지표 : r2 score

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/test.csv

출처 : 5주차 예상문제 1 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/test.csv")
```

```python
train.info()
```

```python
train.describe()
```

## 데이터 전처리 및 분리

```python
X = train.drop(['charges'], axis=1)
y = train['charges']
```

```python
X = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## 모델 검증

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```

## 테스트

```python
test = pd.get_dummies(test)
```

```python
submission = pd.DataFrame(model.predict(test))
```

```python
submission.to_csv("submission_5주차_1.csv", index=False, header=False)
```

# 5주차 예상문제 2

데이터 출처 :https://www.kaggle.com/harlfoxem/housesalesprediction (참고, 데이터 수정)

데이터 설명 : 킹카운티 주거지 가격 예측문제

평가지표 : r2 score

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/test.csv

출처 : 5주차 예상문제 2 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

## 데이터 전처리 및 분리

```python
X = train.drop(['id', 'date', 'zipcode', 'price'], axis=1)
y = train['price']
```

```python
X = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## 모델 검증

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```

## 테스트

```python
test = test.drop(['id', 'date', 'zipcode'], axis=1)
test = pd.get_dummies(test)
```

```python
submission = pd.DataFrame(model.predict(test))
```

```python
submission.to_csv("submission_5주차_2.csv", index=False, header=False)
```
