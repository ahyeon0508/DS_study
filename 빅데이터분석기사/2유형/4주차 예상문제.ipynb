# 4주차 예상문제 1

데이터 출처 :https://www.kaggle.com/teejmahal20/airline-passenger-satisfaction?select=train.csv (참고, 데이터 수정)

데이터 설명 : 비행탑승 경험 만족도 (satisfaction 컬럼 : 'neutral or dissatisfied' or satisfied ) 

문제타입 : 분류유형

평가지표 : accuracy

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/test.csv

출처 : 4주차 예상문제 1 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

## 전처리

```python
X = train.drop(['id', 'satisfaction', 'Arrival Delay in Minutes'], axis=1)
y = train['satisfaction']

X = pd.get_dummies(X)
y = y.map(lambda x: 1 if x == 'satisfied' else 0)
```

## 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## 모델 평가

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_pred, y_test)
```

## 테스트 평가

```python
test = test.drop(['id', 'Arrival Delay in Minutes'], axis=1)
test = pd.get_dummies(test)

model.predict(test)
```

# 4주차 예상문제 2

데이터 설명 : 수질 음용성 여부 (Potablillity 컬럼 : 0 ,1 )

문제타입 : 분류유형

평가지표 : accuracy

trainData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/train.csv

testData url : https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/test.csv​

출처 : 4주차 예상문제 2 (이기적 스터디 카페 / 컴활,정보처리,빅분기) | 작성자 빅분기 실기조교

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/test.csv")
```

```python
train
```

```python
train.info()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

## 전처리

```python
train.median()
```

```python
train = train.fillna(train.median())
test = test.fillna(test.median())
```

## 데이터 분리

```python
X = train.drop(['Potability'], axis=1)
y = train['Potability']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

## 모델 학습

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## 모델 평가

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```

## 테스트

```python
model.predict(test)
```
